#### Overzicht van werkwoorden waarbij DRK/ als object geldt

```python
#prepare list f verbs that have DRK/ as object 
verb_list_objDRK = []
verb_dict_objDRK = {}
for phrase in F.otype.s('phrase'):
    if F.function.v(phrase) == 'Objc':
        obj_woorden = L.d(phrase, 'word')
        for word in obj_woorden:
            if F.lex.v(word) == 'DRK/' and F.language.v(word) == 'Hebrew':
                #find obj phrases wit DRK/
                clause = L.u(word, 'clause')[0]#go up clause and
                phrases = L.d(clause, 'phrase')#go down to find the Pred phrases. Maybe to be extended to other preds
                for phr in phrases:
                    if F.function.v(phr) == 'Pred':
                        #select the verbs
                        pred_woorden = L.d(phr, 'word')
                        for w in pred_woorden:
                            if F.sp.v(w) == 'verb':
                                where = T.sectionFromNode(w)
                                info = [w, where[0], where[1], where[2], F.lex.v(w), F.vt.v(w),F.vs.v(w)]
                                verb_list_objDRK.append(w)
                                verb_dict_objDRK[w] = info
                                
print(len(verb_list_objDRK)) 
pp.pprint(verb_dict_objDRK)       
```

```python
#prepare file of verbs with DRK/ as object
with open(r'mov_pred.csv', "w") as csv_file:

    header = ['book','chapter', 'verse', 'verb_lex', 'verb_tense', 'verb_stem']
    csv_file.write('{}\n'.format(','.join(header)))
    for case in verb_dict_objDRK:
        info = verb_dict_objDRK[case]
        line = [str(element) for element in info]
        csv_file.write('{}\n'.format(','.join(line))) 
  
```

#### Overzicht van werkwoorden waarbij DRK/ als cmplement geldt

```python
#prepare list f verbs that have DRK/ as complement
verb_list_cmplDRK = []
verb_dict_cmplDRK = {}
for phrase in F.otype.s('phrase'):
    if F.function.v(phrase) == 'Cmpl':
        cmpl_woorden = L.d(phrase, 'word')
        for word in cmpl_woorden:
            if F.lex.v(word) == 'DRK/' and F.language.v(word) == 'Hebrew':
                #print(T.sectionFromNode(word))
                clause = L.u(word, 'clause')[0]
                phrases = L.d(clause, 'phrase')
                for phr in phrases:
                    if F.function.v(phr) == 'Pred':
                        #print(T.sectionFromNode(phr))
                        cmpl_woorden = L.d(phr, 'word')
                        for w in cmpl_woorden:
                            if F.sp.v(w) == 'verb':
                                #print(T.sectionFromNode(w))
                                where = T.sectionFromNode(w)
                                info = [w, where[0], where[1], where[2], F.lex.v(w), F.vt.v(w),F.vs.v(w)]
                                verb_list_cmplDRK.append(w)
                                verb_dict_cmplDRK[w] = info
                                
print(len(verb_list_cmplDRK)) 
pp.pprint(verb_dict_cmplDRK)       
```

```python
#prepare file of verbs with DRK/ as cmpl
with open(r'mov_cmpl.csv', "w") as csv_file:

    header = ['book','chapter', 'verse', 'verb_lex', 'verb_tense', 'verb_stem']
    csv_file.write('{}\n'.format(','.join(header)))
    for case in verb_dict_cmplDRK:
        info = verb_dict_cmplDRK[case]
        line = [str(element) for element in info]
        csv_file.write('{}\n'.format(','.join(line))) 
```
