```python
#prepare list of verbs that have DRK/ as complement or object  with preps.
verb_dict_DRK = {}
teller = 0

for word in F.otype.s('word'):
#word = 2821
    if F.lex.v(word) == 'DRK/' and F.language.v(word) == 'Hebrew':
        teller +=1
        clause = L.u(word, 'clause')[0]#going up to clause level
        phrases = L.d(clause, 'phrase')# and going down to all phrases in relevant clause                             
            
        for phr in phrases:#finding relevant phrases
            if F.function.v(phr) == 'Pred':#if predicate phrase
                #print(T.sectionFromNode(phr))
                pred_woorden = L.d(phr, 'word')#go down to word level
                phrase_func = F.function.v(phr)
                for w in pred_woorden:
                    if F.sp.v(w) == 'verb':#find verbs
                        #print(T.sectionFromNode(w))
                        verb_lex = F.lex.v(w)
                        verb_tense = F.vt.v(w)
                        verb_stem = F.vs.v(w)
                                
                                                             

            else:
                if F.function.v(phr) == 'Cmpl':#if predicate phrase
                    #print(T.sectionFromNode(phr))                    
                    cmpl_woorden = L.d(phr, 'word')#go down to word level
                    phrase_func = F.function.v(phr)
                    for w in cmpl_woorden:
                        if F.sp.v(w) == 'verb':# find verbs
                            #print(T.sectionFromNode(w))
                            verb_lex = F.lex.v(w)
                            verb_tense = F.vt.v(w)
                            verb_stem = F.vs.v(w)
                                
                        else:
                            if F.sp.v(w) == 'prep':#find preps
                                #print(F.lex.v(word))
                                #print(T.sectionFromNode(word))
                                prep_lex = F.lex.v(w)
                                    
            
                if F.function.v(phr) == 'Objc':#if object phrase
                    #print(T.sectionFromNode(phr))                    
                    objc_woorden = L.d(phr, 'word')
                    phrase_fun = F.function.v(phr)
                    for w in objc_woorden:
                        if F.sp.v(w) == 'verb':#find verbs
                            #print(T.sectionFromNode(w))
                            verb_lex = F.lex.v(w)
                            verb_tense = F.vt.v(w)
                            verb_stem = F.vs.v(w)
                                   
                                

    where = T.sectionFromNode(word)
    info = [word, where[0], where[1], where[2], phrase_func, verb_lex, verb_tense,verb_stem, prep_lex]
    verb_dict_DRK[w] = info      
                                
print(teller)       
pp.pprint(verb_dict_DRK)
```
